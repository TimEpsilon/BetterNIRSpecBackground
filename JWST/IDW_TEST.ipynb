```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0,256,10)
y = np.random.randint(0,128,10)
z = np.random.random(10)*10
X,Y = np.meshgrid(range(256),range(128))

xy_ravel = np.column_stack((X.ravel(),Y.ravel()))
x_ravel = xy_ravel[:, 0]
y_ravel = xy_ravel[:, 1]

print(f"x = {x.shape}")
print(f"y = {y.shape}")
print(f"X = {X.shape}")
print(f"Y = {Y.shape}")
print(f"X_ravel = {x_ravel.shape}")
print(f"Y_ravel = {y_ravel.shape}")
```

```python
for p in range(10): 
    X1, X2 = np.meshgrid(x,x_ravel)
    Y1, Y2 = np.meshgrid(y,y_ravel)

    d = ((X1-X2)**2 + (Y1-Y2)**2).T
    w = d**(-p/2)

    w_ui_sum = z[:, None]*w
    w_ui_sum = w_ui_sum.sum(axis=0)

    wsum = w.sum(axis=0)

    result = w_ui_sum / wsum

    result = result.reshape(np.shape(X))
    result[y,x] = z
    
    plt.figure()
    plt.imshow(result)
    plt.colorbar()
    plt.title(f"p = {p}")
    plt.scatter(x,y,color='red',marker='+')
    plt.show()


```
